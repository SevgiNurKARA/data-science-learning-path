## 3. Küçük Veri Seti Oluşturma (NumPy ile)


```python
# Öğrenci notları veri seti oluşturma
np.random.seed(42)  # Tekrarlanabilirlik için

# 20 öğrenci için veri oluştur
n_students = 20

# Öğrenci ID'leri
student_ids = np.arange(1, n_students + 1)

# Matematik notları (50-100 arası)
math_scores = np.random.randint(50, 101, n_students)

# Fizik notları (45-100 arası)
physics_scores = np.random.randint(45, 101, n_students)

# Kimya notları (40-100 arası)
chemistry_scores = np.random.randint(40, 101, n_students)

# Yaşlar (18-25 arası)
ages = np.random.randint(18, 26, n_students)

print("Öğrenci ID'leri:", student_ids[:5])
print("Matematik notları:", math_scores[:5])
print("Fizik notları:", physics_scores[:5])
print("Kimya notları:", chemistry_scores[:5])
print("Yaşlar:", ages[:5])
```

## 4. NumPy Array'lerini Pandas DataFrame'e Dönüştürme


```python
# NumPy array'lerinden DataFrame oluşturma
df = pd.DataFrame({
    'Student_ID': student_ids,
    'Age': ages,
    'Math': math_scores,
    'Physics': physics_scores,
    'Chemistry': chemistry_scores
})

print("DataFrame oluşturuldu:")
df.head(10)
```

## 5. Pandas ile Veri Analizi (NumPy fonksiyonları kullanarak)


```python
# Ortalama not hesaplama (NumPy mean kullanarak)
df['Average'] = np.mean([df['Math'], df['Physics'], df['Chemistry']], axis=0)
df.head()
```

```python
# Daha iyi yöntem: Pandas ile
df['Average'] = df[['Math', 'Physics', 'Chemistry']].mean(axis=1)
df.head()
```

```python
# NumPy where ile koşullu sütun oluşturma
# Ortalama 70'ten büyükse 'Başarılı', değilse 'Başarısız'
df['Status'] = np.where(df['Average'] >= 70, 'Başarılı', 'Başarısız')
df.head(10)
```

```python
# NumPy ile kategorik değerler oluşturma
# Ortalamaya göre harf notu
conditions = [
    df['Average'] >= 90,
    (df['Average'] >= 80) & (df['Average'] < 90),
    (df['Average'] >= 70) & (df['Average'] < 80),
    (df['Average'] >= 60) & (df['Average'] < 70),
    df['Average'] < 60
]
choices = ['A', 'B', 'C', 'D', 'F']
# np.select için koşulları numpy array'e dönüştür ve default değer belirt
df['Grade'] = np.select(conditions, choices, default='F')
df.head(10)
```

## 6. NumPy Array İşlemleri ile Veri Manipülasyonu


```python
# DataFrame'i NumPy array'e dönüştürme
# Sadece sayısal sütunlar
numeric_data = df[['Math', 'Physics', 'Chemistry']].values
print("NumPy array shape:", numeric_data.shape)
print("\nİlk 5 satır:")
print(numeric_data[:5])
```

```python
# NumPy ile array üzerinde işlemler
# Her öğrencinin notlarını normalize etme (0-1 arası)
normalized_scores = (numeric_data - numeric_data.min(axis=0)) / (numeric_data.max(axis=0) - numeric_data.min(axis=0))
print("Normalize edilmiş notlar (ilk 5 satır):")
print(normalized_scores[:5])
```

```python
# Normalize edilmiş değerleri DataFrame'e ekleme
df['Math_Normalized'] = normalized_scores[:, 0]
df['Physics_Normalized'] = normalized_scores[:, 1]
df['Chemistry_Normalized'] = normalized_scores[:, 2]
df[['Math', 'Math_Normalized', 'Physics', 'Physics_Normalized']].head()
```

## 7. İstatistiksel Analiz (NumPy + Pandas)


```python
# NumPy ile istatistiksel özet
print("=== NumPy ile İstatistikler ===")
print(f"Matematik - Ortalama: {np.mean(df['Math']):.2f}")
print(f"Matematik - Medyan: {np.median(df['Math']):.2f}")
print(f"Matematik - Standart Sapma: {np.std(df['Math']):.2f}")
print(f"Matematik - Min: {np.min(df['Math'])}")
print(f"Matematik - Max: {np.max(df['Math'])}")
print(f"Matematik - 25. Percentile: {np.percentile(df['Math'], 25):.2f}")
print(f"Matematik - 75. Percentile: {np.percentile(df['Math'], 75):.2f}")
```

## 9. NumPy Array İşlemleri ile Gelişmiş Analiz


```python
# NumPy ile vektörleştirilmiş işlemler
# Her ders için z-score hesaplama
math_zscore = (df['Math'] - np.mean(df['Math'])) / np.std(df['Math'])
physics_zscore = (df['Physics'] - np.mean(df['Physics'])) / np.std(df['Physics'])
chemistry_zscore = (df['Chemistry'] - np.mean(df['Chemistry'])) / np.std(df['Chemistry'])

df['Math_ZScore'] = math_zscore
df['Physics_ZScore'] = physics_zscore
df['Chemistry_ZScore'] = chemistry_zscore

print("Z-Score değerleri (ilk 5 öğrenci):")
df[['Student_ID', 'Math', 'Math_ZScore', 'Physics', 'Physics_ZScore']].head()
```

```python
# NumPy ile matris çarpımı örneği
# Ağırlıklı ortalama hesaplama
scores_matrix = df[['Math', 'Physics', 'Chemistry']].values
weights = np.array([0.4, 0.35, 0.25])  # Math, Physics, Chemistry ağırlıkları
weighted_avg = np.dot(scores_matrix, weights)

df['Weighted_Average'] = weighted_avg
print("Ağırlıklı Ortalama (Math:0.4, Physics:0.35, Chemistry:0.25):")
df[['Student_ID', 'Math', 'Physics', 'Chemistry', 'Average', 'Weighted_Average']].head(10)
```

```python
# NumPy ile argmax/argmin - en yüksek/düşük notun hangi derste olduğunu bulma
best_subject_idx = np.argmax(scores_matrix, axis=1)
worst_subject_idx = np.argmin(scores_matrix, axis=1)

subjects = ['Math', 'Physics', 'Chemistry']
df['Best_Subject'] = [subjects[idx] for idx in best_subject_idx]
df['Worst_Subject'] = [subjects[idx] for idx in worst_subject_idx]

print("Her öğrencinin en iyi ve en kötü dersi:")
df[['Student_ID', 'Math', 'Physics', 'Chemistry', 'Best_Subject', 'Worst_Subject']].head(10)
```

## 18. Performans Karşılaştırması (NumPy vs Python List)


```python
# Performans karşılaştırması
import time

# Büyük veri seti oluştur
large_size = 10000000  # Daha büyük veri seti (10 milyon)
python_list = list(range(large_size))
numpy_array = np.arange(large_size)

# Python list ile toplama (birden fazla kez çalıştırarak daha doğru ölçüm)
start = time.perf_counter()  # perf_counter daha hassas
for _ in range(10):  # 10 kez çalıştır
    sum_python = sum(python_list)
time_python = (time.perf_counter() - start) / 10  # Ortalama

# NumPy ile toplama (birden fazla kez çalıştırarak daha doğru ölçüm)
start = time.perf_counter()
for _ in range(10):  # 10 kez çalıştır
    sum_numpy = np.sum(numpy_array)
time_numpy = (time.perf_counter() - start) / 10  # Ortalama

print(f"Python List Toplama: {time_python:.6f} saniye")
print(f"NumPy Array Toplama: {time_numpy:.6f} saniye")

# Sıfıra bölme kontrolü
if time_numpy > 0:
    speedup = time_python / time_numpy
    print(f"NumPy {speedup:.2f}x daha hızlı!")
else:
    print("NumPy çok hızlı! Zaman ölçümü çok küçük (0'a yakın).")
    print(f"Python List: {time_python:.9f} saniye")
    print(f"NumPy Array: {time_numpy:.9f} saniye (veya daha küçük)")
```

## 19. Gerçek Dünya Senaryosu: Satış Verisi Analizi


```python
# Satış verisi oluşturma
np.random.seed(42)
n_days = 30
dates = pd.date_range('2024-01-01', periods=n_days, freq='D')

sales_data = pd.DataFrame({
    'Date': dates,
    'Product_A': np.random.randint(10, 50, n_days),
    'Product_B': np.random.randint(15, 60, n_days),
    'Product_C': np.random.randint(5, 40, n_days),
    'Price_A': np.random.uniform(10, 20, n_days),
    'Price_B': np.random.uniform(15, 25, n_days),
    'Price_C': np.random.uniform(8, 18, n_days)
})

# Gelir hesaplama (NumPy ile)
sales_data['Revenue_A'] = sales_data['Product_A'] * sales_data['Price_A']
sales_data['Revenue_B'] = sales_data['Product_B'] * sales_data['Price_B']
sales_data['Revenue_C'] = sales_data['Product_C'] * sales_data['Price_C']
sales_data['Total_Revenue'] = np.sum([sales_data['Revenue_A'], 
                                      sales_data['Revenue_B'], 
                                      sales_data['Revenue_C']], axis=0)

print("Satış Verisi:")
sales_data.head(10)
```

```python
# Günlük toplam gelir analizi
print("=== SATIŞ ANALİZİ ===")
print(f"Toplam Gelir: ${sales_data['Total_Revenue'].sum():.2f}")
print(f"Günlük Ortalama Gelir: ${sales_data['Total_Revenue'].mean():.2f}")
print(f"En Yüksek Günlük Gelir: ${sales_data['Total_Revenue'].max():.2f}")
print(f"En Düşük Günlük Gelir: ${sales_data['Total_Revenue'].min():.2f}")

# En karlı gün
best_day = sales_data.loc[sales_data['Total_Revenue'].idxmax()]
print(f"\nEn Karlı Gün: {best_day['Date'].strftime('%Y-%m-%d')}")
print(f"Gelir: ${best_day['Total_Revenue']:.2f}")
```

```python
# Ürün bazında analiz
product_revenue = {
    'Product_A': sales_data['Revenue_A'].sum(),
    'Product_B': sales_data['Revenue_B'].sum(),
    'Product_C': sales_data['Revenue_C'].sum()
}

print("Ürün Bazında Toplam Gelir:")
for product, revenue in sorted(product_revenue.items(), key=lambda x: x[1], reverse=True):
    percentage = (revenue / sales_data['Total_Revenue'].sum()) * 100
    print(f"{product}: ${revenue:.2f} (%{percentage:.1f})")
```

```python
# Gelir trendi görselleştirme
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(sales_data['Date'], sales_data['Total_Revenue'], marker='o', linewidth=2)
plt.title('Günlük Toplam Gelir Trendi')
plt.xlabel('Tarih')
plt.ylabel('Gelir ($)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
products = ['Product_A', 'Product_B', 'Product_C']
revenues = [sales_data[f'Revenue_{p[-1]}'].sum() for p in products]
plt.bar(products, revenues, color=['blue', 'green', 'red'], alpha=0.7)
plt.title('Ürün Bazında Toplam Gelir')
plt.ylabel('Gelir ($)')
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
# NumPy apply_along_axis - belirli eksen boyunca fonksiyon uygulama
# Her öğrencinin notlarının standart sapmasını hesapla
scores_array = df[['Math', 'Physics', 'Chemistry']].values
std_per_student = np.apply_along_axis(np.std, axis=1, arr=scores_array)
df['Score_Std'] = std_per_student
print("Her öğrencinin notlarının standart sapması:")
df[['Student_ID', 'Math', 'Physics', 'Chemistry', 'Score_Std']].head(10)
```

```python
# NumPy bincount - belirli aralıktaki değerlerin sayısı
# Not aralıklarına göre öğrenci sayısı
score_bins = np.bincount(df['Math'] // 10)  # 10'ar 10'ar grupla
print("Not aralıklarına göre öğrenci sayısı:")
for i, count in enumerate(score_bins):
    if count > 0:
        print(f"{i*10}-{(i+1)*10-1} arası: {count} öğrenci")
```
