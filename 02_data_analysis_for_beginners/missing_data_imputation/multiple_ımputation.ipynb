```python
import pandas as pd

data = pd.read_csv("data_resampled.csv")
data
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=3)
data_knn = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
data_knn
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer(max_iter=10, random_state=0)
df_mi = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
df_mi
```

```python
from sklearn.experimental import enable_iterative_imputer
import numpy as np
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
imputer = IterativeImputer(estimator=RandomForestRegressor(), random_state=0)

df_rfi = pd.DataFrame(imputer.fit_transform(data.select_dtypes(include=[np.number])),
                      columns=data.select_dtypes(include=[np.number]).columns)
df_rfi
```

```python
from sklearn.linear_model import LinearRegression

# Önce Age ve Testosterone değerlerinin her ikisinin de mevcut olduğu verileri al
df_complete = data.dropna(subset=['Age', 'Testosterone_Level(ng/dL)'])

# Sadece Age değeri olan ama Testosterone değeri eksik olan verileri al
df_to_predict = data[
    data['Age'].notna() & 
    data['Testosterone_Level(ng/dL)'].isna()
]

if len(df_to_predict) > 0:  # Tahmin edilecek veri varsa
    # Modeli eğit
    model = LinearRegression()
    X_train = df_complete[['Age']]
    y_train = df_complete['Testosterone_Level(ng/dL)']
    model.fit(X_train, y_train)

    # Tahmin yap
    X_test = df_to_predict[['Age']]
    predicted_values = model.predict(X_test)

    # Tahminleri yerleştir
    data.loc[df_to_predict.index, 'Testosterone_Level(ng/dL)'] = predicted_values

data
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

mice_imputer = IterativeImputer(max_iter=10, random_state=0)

numeric_columns = data.select_dtypes(include=[np.number]).columns
df_imputed_mice = pd.DataFrame(mice_imputer.fit_transform(data[numeric_columns]), 
                               columns=numeric_columns)
df_imputed_mice
```
