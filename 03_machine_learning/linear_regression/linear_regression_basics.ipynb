# Linear Regression - Temel Kavramlar

Bu notebook, Linear Regression algoritmasının temel kavramlarını ve uygulamasını içerir.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_regression

plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
%matplotlib inline
```

## 1. Basit Linear Regression


```python
# Veri seti oluştur
X, y = make_regression(n_samples=1000, n_features=1, noise=20, random_state=42)

# Model eğit
model = LinearRegression()
model.fit(X, y)

# Tahmin
y_pred = model.predict(X)

# Metrikler
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print(f"MAE:  {mae:.4f}")
print(f"MSE:  {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²:   {r2:.4f}")
print(f"\nKatsayı: {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")

# Görselleştir
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.5, label='Gerçek Değerler')
plt.plot(X, y_pred, color='red', linewidth=2, label='Tahmin')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 2. Multiple Linear Regression


```python
# Çoklu feature ile veri seti
X_multi, y_multi = make_regression(n_samples=1000, n_features=5, noise=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_multi, y_multi, test_size=0.2, random_state=42)

# Model eğit
model_multi = LinearRegression()
model_multi.fit(X_train, y_train)

# Tahmin
y_pred_multi = model_multi.predict(X_test)

# Metrikler
print("Multiple Linear Regression Metrikleri:")
print(f"MAE:  {mean_absolute_error(y_test, y_pred_multi):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred_multi)):.4f}")
print(f"R²:   {r2_score(y_test, y_pred_multi):.4f}")
print(f"\nKatsayılar: {model_multi.coef_}")
print(f"Intercept: {model_multi.intercept_:.4f}")
```

## 3. Regularization: Ridge, Lasso, Elastic Net


```python
# Feature scaling (regularization için önemli)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Ridge Regression (L2)
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)
y_pred_ridge = ridge.predict(X_test_scaled)

# Lasso Regression (L1)
lasso = Lasso(alpha=1.0)
lasso.fit(X_train_scaled, y_train)
y_pred_lasso = lasso.predict(X_test_scaled)

# Elastic Net (L1 + L2)
elastic = ElasticNet(alpha=1.0, l1_ratio=0.5)
elastic.fit(X_train_scaled, y_train)
y_pred_elastic = elastic.predict(X_test_scaled)

# Karşılaştırma
results = pd.DataFrame({
    'Model': ['Linear', 'Ridge', 'Lasso', 'Elastic Net'],
    'R²': [
        r2_score(y_test, model_multi.predict(X_test)),
        r2_score(y_test, y_pred_ridge),
        r2_score(y_test, y_pred_lasso),
        r2_score(y_test, y_pred_elastic)
    ],
    'RMSE': [
        np.sqrt(mean_squared_error(y_test, model_multi.predict(X_test))),
        np.sqrt(mean_squared_error(y_test, y_pred_ridge)),
        np.sqrt(mean_squared_error(y_test, y_pred_lasso)),
        np.sqrt(mean_squared_error(y_test, y_pred_elastic))
    ]
})

print("Regularization Yöntemlerinin Karşılaştırması:")
print(results.round(4))
```
