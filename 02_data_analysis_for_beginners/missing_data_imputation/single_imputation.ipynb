```python
import pandas as pd

data = pd.read_csv("data_resampled.csv")
data
```

```python
data.describe()
```

İnplace parametresi "true" dönmedikçe orijinal veri setinde değişiklik olmayacaktır.

```python
#Mean ile doldurma
df_mean=data.fillna(data.mean())
df_mean
```

```python
df_mode=data.fillna(data.mode())
df_mode
```

```python
df_median=data.fillna(data.median())
df_median
```

Tekli yaklaşım methodu istatiksel temele dayanır. İstediğiniz istatiksel veri ile eksik verileri tamamlamanız mümkündür. Hatta
fillna() methodunu kullanarak tüm eksik değerleri belirli bir sayıyla bile tamamlayabilirsiniz.
Bu metottan max verim almak için veri setinize dikkat etmelisiniz.
