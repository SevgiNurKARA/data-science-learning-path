```python
# Yeni sütun ekleme
movies['Score_Category'] = movies['Score'].apply(lambda x: 'High' if x > 8 else 'Low')
movies.head()
```

```python
# value_counts() - kategorik değerlerin sayısını gösterir
# Örnek: Her yılda kaç film var?
movies['Year'].value_counts()
```

```python
# unique() ve nunique() - benzersiz değerler
print("Benzersiz yıl sayısı:", movies['Year'].nunique())
print("Benzersiz yıllar:", movies['Year'].unique()[:10])  # İlk 10 yıl
```

```python
# String işlemleri - .str erişicisi ile
# Film isimlerinin uzunluğunu hesaplama
movies['Title_Length'] = movies['Movie Title'].str.len()
movies[['Movie Title', 'Title_Length']].head()
```

```python
# String içinde arama - belirli kelimeyi içeren filmler
movies[movies['Movie Title'].str.contains('Dragon', case=False)]
```

```python
# Aggregation (toplulaştırma) işlemleri
# Yıllara göre ortalama score
movies.groupby('Year')['Score'].mean().sort_values(ascending=False).head(10)
```

```python
# Birden fazla aggregation fonksiyonu
movies.groupby('Year')['Score'].agg(['mean', 'min', 'max', 'count']).head(10)
```

```python
# query() metodu ile filtreleme (daha okunabilir)
movies.query('Score > 95 and Year > 2010')
```

```python
# replace() - değer değiştirme
# Örnek: Score'u kategorilere çevirme
movies['Score_Category'] = movies['Score'].replace(
    {range(90, 101): 'Excellent', 
     range(80, 90): 'Good', 
     range(70, 80): 'Average'}, 
    regex=False
)
# Daha basit yöntem:
movies['Score_Category'] = pd.cut(movies['Score'], 
                                  bins=[0, 80, 90, 100], 
                                  labels=['Good', 'Very Good', 'Excellent'])
movies[['Movie Title', 'Score', 'Score_Category']].head()
```

```python
# pivot_table - çapraz tablo oluşturma
# Yıllara göre score dağılımı (eğer kategori varsa)
# Örnek: Yıllara göre ortalama score pivot tablosu
movies.pivot_table(values='Score', index='Year', aggfunc='mean').head(10)
```

```python
# concat() - DataFrame'leri birleştirme
# Örnek: İlk 5 ve son 5 filmi birleştirme
top_5 = movies.head(5)
bottom_5 = movies.tail(5)
combined = pd.concat([top_5, bottom_5], ignore_index=True)
combined[['Movie Title', 'Score']]
```

```python
# merge() - iki DataFrame'i birleştirme
# Örnek: Yıllara göre istatistikler oluşturup birleştirme
year_stats = movies.groupby('Year').agg({
    'Score': ['mean', 'count']
}).reset_index()
year_stats.columns = ['Year', 'Avg_Score', 'Movie_Count']
# Orijinal DataFrame ile birleştirme
movies_with_stats = movies.merge(year_stats, on='Year', how='left')
movies_with_stats[['Movie Title', 'Year', 'Score', 'Avg_Score', 'Movie_Count']].head()
```

```python
# nlargest() ve nsmallest() - en büyük/küçük değerler
# En yüksek score'a sahip 5 film
movies.nlargest(5, 'Score')[['Movie Title', 'Year', 'Score']]
```

```python
# duplicated() - tekrar eden kayıtları bulma
# Tekrar eden yılları kontrol etme (bu örnekte her film farklı)
print("Tekrar eden kayıt sayısı:", movies.duplicated().sum())
# drop_duplicates() - tekrarları silme
# movies.drop_duplicates(subset=['Year'], keep='first')
```

```python
# reset_index() - index'i sıfırlama
# Sıralama yaptıktan sonra index'i düzeltme
sorted_movies = movies.sort_values('Score', ascending=False).reset_index(drop=True)
sorted_movies.head()
```

```python
# İleri seviye filtreleme - birden fazla koşul
# Score'u 90'dan büyük VE 2010'dan sonra çıkan filmler
high_score_recent = movies[(movies['Score'] > 90) & (movies['Year'] > 2010)]
high_score_recent[['Movie Title', 'Year', 'Score']]
```

```python
# isin() - belirli değerler listesinde olanları filtreleme
# Belirli yıllardaki filmler
specific_years = movies[movies['Year'].isin([2010, 2020, 2021])]
specific_years[['Movie Title', 'Year', 'Score']].sort_values('Year')
```
